```python
from tabulate import tabulate
import keyboard


produtos = []


def cadastrar_produto():
    print("\nCadastro de Produto:")
    
    nome = input("Nome do Produto: ")
    descricao = input("Descrição do Produto: ")
    valor = float(input("Valor do Produto: R$ "))
    disponivel = input("Disponível para venda (sim/não): ").strip().lower()
    
    
    while disponivel not in ["sim", "não"]:
        print("Opção inválida. Digite 'sim' ou 'não'.")
        disponivel = input("Disponível para venda (sim/não): ").strip().lower()

    
    produto = {
        "nome": nome,
        "descricao": descricao,
        "valor": valor,
        "disponivel": disponivel
    }
    
    
    produtos.append(produto)
    print("\nProduto cadastrado com sucesso!\n")


def listar_produtos():
    if len(produtos) == 0:
        print("Nenhum produto cadastrado.\n")
        return
    
    
    produtos.sort(key=lambda p: p["valor"])

    
    cabecalho = ["Nome", "Descrição", "Valor", "Disponível"]
    
    
    dados = [(produto["nome"], produto["descricao"], f"R$ {produto['valor']:.2f}", produto["disponivel"]) for produto in produtos]
    
    
    print("\nLista de Produtos Cadastrados:")
    print(tabulate(dados, headers=cabecalho, tablefmt="fancy_grid", numalign="center"))


def menu():
    while True:
        print("Escolha uma opção:")
        print("1. Cadastrar produto")
        print("2. Listar produtos")
        print("3. Sair")
        
        opcao = input("Digite o número da opção: ")

        
        if opcao == "1":
            cadastrar_produto()
        elif opcao == "2":
            listar_produtos()
        elif opcao == "3" or keyboard.is_pressed('esc'): 
            print("Saindo do sistema...")
            break
        else:
            print("\nOpção inválida. Tente novamente.\n")

menu()
```
